```python
##Import Modules##
import numpy as np
import pandas as pd
import datetime as dt
from pandas_datareader import data
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
##Reading Data from Yahoo finance##
tickers= ['MSFT','JPM','SBUX','KSS']
start_date=dt.datetime(2010,1,1)
end_date=dt.datetime(2018,12,31)
stock_data=data.DataReader(tickers,'yahoo',start_date,end_date)['Adj Close']
stock_data
```

```python
### Plotting the stock data
stock_data.plot()
plt.show()
```

```python
###calculating Stock daily returns
stock_returns=((stock_data)/stock_data.shift(1))-1
stock_returns
```

```python
####Plotting the returns using matplotlib
stock_returns.plot()
plt.title('Stock Returns for JPM MSFT KSS SBUX')
plt.show()
```

```python
###Assigning wieghtage using numpy arrays
weight=np.array([0.25,0.25,0.25,0.25])
weight
```

```python
###calculating stock average annual returns
average_returns=stock_returns.mean()*250
average_returns
```

```python
###Portfolio Return Calculation,weight*returns
Portfolio_return=np.dot(average_returns,weight)
Portfolio_return                    
```

```python
ROR_on_Portfolio=str(round(Portfolio_return,4)*100)+'%'
```
